```python
# Number of Numeric and String Type Columns

loans = pd.read_csv("loans_2007.csv",chunksize=3000)

for chunk in loans:
    num_types = chunk.select_dtypes(include=['float64'])
    str_types = chunk.select_dtypes(include=['object'])
    len_num = len(num_types.columns)
    str_num = len(str_types.columns)

print("Number of Numeric Columns:",len_num)
print("Number of String Columns:",str_num)
```

```python
# Unique values in String Columns

unique_strings = []

for col in str_types:
    str_uniq = chunk[col].nunique()
    if str_uniq / len(chunk[col]) < 0.50:
        unique_strings.append(col)
    print("string column:",col,"\n# of unique values:",str_uniq)
print("\nColumns with fewer than 50% unique values:",len(unique_strings),",",len(unique_strings)/str_num)
```

```python
# Float Columns with no Missing Values

nona_cols = []

for col_name in num_types:
    if chunk[col_name].notnull().all():
        nona_cols.append(col_name)
    na_cols = num_types.isnull().sum()
print("Float Columns with no missing values:",nona_cols,"\n")
print("Number of Missing Values in Float Columns:")
print(na_cols) 
```

```python
# Total memory usage across all of the chunks

total_mem = 0
for chunk in pd.read_csv("loans_2007.csv",chunksize=3000):
    total_mem += chunk.memory_usage(deep=True).sum()
print("Total Memory Usage:",total_mem/1048576,"MB")
```

```python
"""Clean columns and convert to more space efficient types"""

total_mem_ft = 0

chunk_iter = pd.read_csv("loans_2007.csv",chunksize=3000)

for chunk in chunk_iter:
    
    #convert id to float
    chunk['id'] = chunk['id'].apply(pd.to_numeric, errors='coerce')

    # convert term to float
    chunk["term"] = chunk["term"].str.replace("months","")
    chunk["term"] = chunk["term"].astype('float')

    # convert int_rate to float
    chunk["int_rate"] = chunk["int_rate"].str.replace("%","")
    chunk["int_rate"] = chunk["int_rate"].astype('float')

    # convert revol_util to float
    chunk["revol_util"] = chunk["revol_util"].str.replace("%","")
    chunk["revol_util"] = chunk["revol_util"].astype('float')
    
    chunk["grade"] = chunk["grade"].astype("category")
    chunk["sub_grade"] = chunk["sub_grade"].astype("category")
    chunk["home_ownership"] = chunk["home_ownership"].astype("category")
    chunk["verification_status"] = chunk["verification_status"].astype("category")
    chunk["pymnt_plan"] = chunk["pymnt_plan"].astype("category")
    
    # Since no columns have any missing values, 
    # I made the judgement call to drop the 2 rows that had missing
    # values in every column, so I would be able to convert some
    # columns to int type

    chunk.dropna(subset=['id'],axis=0,inplace=True)
    
    
    # convert id from float to int32
    chunk["id"] = chunk["id"].astype('int32')
    
    # convert member_id from float to int32
    chunk["member_id"] = chunk["member_id"].astype('int32')
    
    # convert term from float to int16
    chunk["term"] = chunk["term"].astype('int16')
    
    # convert revol_util from float64 to float16
    chunk["revol_util"] = chunk["revol_util"].astype('float16')
    
    # convert int_rate from float64 to float16
    chunk["int_rate"] = chunk["int_rate"].astype('float16')
    
    # add up the memory usage across each chunk
    total_mem_ft += chunk.memory_usage(deep=True).sum()
  
# compare the total memory usage now with before
print("Total Memory Usage:",total_mem_ft/1048576,"MB")
```
